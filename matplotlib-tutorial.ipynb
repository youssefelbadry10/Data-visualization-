```python
%matplotlib inline 
plt.figure(figsize=(8,6))
company =["Google", "AMZN", "MSFT", "FB"]
revenue =[90,136,85,27]
profit =[40,7,34,12]
xpos = np.arange(len(company))

plt.bar(company, revenue, width=0.4 , label= "Revenue")
plt.bar(xpos-0.4, profit, width=0.4 , label= "Revenue")

plt.xticks(xpos, company)
plt.ylabel("Revenue(Bln)")
plt.title("US Technology Stocks")

plt.legend()
plt.show()
```

```python
#make four graphs(2, 2)
x = np.arange(100)
fig , axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sin wave ")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cos wave ")

axs[1, 0].plot(x, np.log(x))
axs[1, 0].set_title("Log wave ")

axs[1, 1].plot(x, np.random.random(100))
axs[1, 1].set_title("Random wave ")

plt.show()

```

```python
# 3D Plots 
plt.rcParams["figure.figsize"]= (8,6)
np.random.seed(31)
mu = 3
n = 50

x = np.random.normal(mu, 1, size=n)
y = np.random.normal(mu, 1, size=n)
z = np.random.normal(mu, 1, size=n)

axis = plt.axes(projection = "3d")
axis.scatter3D(x, y, z)
axis.set_xlabel("x")
axis.set_ylabel("y")
axis.set_zlabel("z")
axis.view_init(45, 215)
```

```python
axis = plt.axes(projection = "3d")
x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)
axis.plot(x, y, z)
axis.set_title("3D plot ")
axis.set_xlabel("test")
plt.show()
```

```python
axis = plt.axes(projection = "3d")
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X,Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)


axis.plot_surface(X, Y, Z)
axis.set_title("3D plot ")
axis.set_xlabel("test")
plt.show()
```

```python
axis = plt.axes(projection = "3d")
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X,Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)


axis.plot_surface(X, Y, Z, cmap = "Spectral")
axis.set_title("3D plot ")
axis.set_xlabel("test")
plt.show()
```
