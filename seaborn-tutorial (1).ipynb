```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
titanic = sns.load_dataset("titanic")
```

```python
tips
```

```python
#scatterplot 
sns.scatterplot(x="tip", y="total_bill" , data=tips)
```

```python
#add an extra column (day)
sns.scatterplot(x="tip", y="total_bill" , data=tips , hue = "day")
```

```python
#you can mange the size of it 
sns.scatterplot(x="tip", y="total_bill" , data=tips , hue = "day", size ="size")
```

```python
#you can manage the color by mention two letters of any color in (palette = "")
sns.scatterplot(x="tip", y="total_bill" , data=tips , hue = "day", size ="size", palette = "YlGnBu")
```

```python
#histographs (kde is the line if you want to put it or no)(bins = number of bins you want )
sns.histplot(tips["tip"], kde = True , bins = 15)
```

```python
#distripution plot
sns.displot(tips["tip"], kde = True , bins = 15, color = "red")
```

```python
#bar plot 
sns.barplot(x="tip", y="sex" , data=tips, palette = "YlGnBu")
```

# **Box plots**

```python
#box plot 
sns.boxplot(x="day", y="tip", data=tips,  palette="YlGnBu")
```

```python
sns.boxplot(x="day", y="tip", data=tips, hue= "sex", palette="YlGnBu")
```

# **strip plots**

```python
#strip plot 
sns.stripplot(x="day", y="tip", data=tips  )
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex" , palette = "YlGnBu" )
```

```python
# you use dodge to separate the colors 
sns.stripplot(x="day", y="tip", data=tips, hue="sex" , palette = "YlGnBu", dodge = True )
```

# **Joint plots**

```python
#joint plot 
# it make a scatter plot for 2 columns and show you the distripution of each one 
sns.jointplot(x="total_bill", y="tip", data=tips )
```

```python
#reg prefer for a regretion line 
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde", fill = True)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde",fill =True, cmap="YlGnBu")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="resid")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex", cmap="YlGnBu")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hist")
```

# **Pair plots**


# Pair plots will plot scatterplots for numerical columns and histograms for the diagonal where a variable is plotted against itself. It's a quick way to visualize relationships between multiple variables in your dataset.

```python
sns.pairplot(tips, hue='sex', kind='scatter', diag_kind='kde')
```
